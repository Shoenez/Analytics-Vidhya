# Health Analytics Naïve Bayes

## Loading in Data

```python
import pandas as pd, numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('train_aj.csv')
test = pd.read_csv('test_v2.csv')
subm = pd.read_csv('s_sub.csv')
```

## Processing data

```python
train_new = pd.read_csv('train_new.csv')
test_new = pd.read_csv('test_new.csv')
train_new2 = pd.read_csv('train_new2.csv')
test_new2 = pd.read_csv('test_new2.csv')
```

```python
train_x = train_new.drop(["stroke","id"],axis=1)
train_y = train_new["stroke"]
test_x = test_new.iloc[:,1:]
```

```python
train_x2 = train_new2.drop(["stroke","id"],axis=1)
train_y2 = train_new2["stroke"]
test_x2 = test_new2.iloc[:,1:]
```

## Building Model

```python
def bnb(y_i, y):
    p = train_x[y==y_i].sum(0)
    return (p+1) / ((y==y_i).sum()+1)
```

```python
def get_mdl(y):
    y = y.values
    r = np.log(bnb(1,y) / bnb(0,y))
    m = LogisticRegression(C=4, dual=True)
    x_nb = train_x.multiply(r)
    return m.fit(x_nb, y), r
```

```python
preds = np.zeros((len(test_x), 2))
m,r = get_mdl(train_y)
preds[:,1] = m.predict_proba(test_x.multiply(r))[:,1]
```

## Building model 2

```python
def bnb2(y_i, y):
    p = train_x2[y==y_i].sum(0)
    return (p+1) / ((y==y_i).sum()+1)
```

```python
def get_mdl2(y):
    y = y.values
    r = np.log(bnb2(1,y) / bnb2(0,y))
    m = LogisticRegression(C=4, dual=True)
    x_nb = train_x2.multiply(r)
    return m.fit(x_nb, y), r
```

```python
preds2 = np.zeros((len(test_x2), 2))
m,r = get_mdl2(train_y2)
preds2[:,1] = m.predict_proba(test_x2.multiply(r))[:,1]
```

## Submission

```python
preds[:,0] = test.iloc[:,0]
submis = pd.DataFrame(preds, columns = ["id","stroke"])
pd.concat([submis], axis=1)

submis.to_csv('nbayes.csv', index=False)
```

```python
preds2[:,0] = test.iloc[:,0]
submis2 = pd.DataFrame(preds2, columns = ["id","stroke"])
pd.concat([submis2], axis=1)

submis2.to_csv('nbayes2.csv', index=False)
```
